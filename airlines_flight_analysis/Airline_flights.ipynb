```python
db['arrival_time'].value_counts()
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1, 2, 1)
plt.bar(
    db['departure_time'].value_counts().index,
    db['departure_time'].value_counts().values,
    color='y')

plt.title('Departure Time')
plt.xlabel('Departure Time')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.bar(
    db['arrival_time'].value_counts().index,
    db['arrival_time'].value_counts().values,
    color='g')

plt.title('Arrival Time')
plt.xlabel('Arrival Time')
plt.ylabel('Frequency')

plt.show()
```

# *Does price varies with airlines?*

```python
db['class'].value_counts()
```

```python
sns.catplot(x='airline' , y='price' , kind = 'bar' , palette = 'rocket' , data = db , hue = 'class')
plt.show()
```

# *Does ticket price changeing according to the time of departure and arrival ?*

```python
db.head()
```

```python
db.groupby(['departure_time','arrival_time'])['price'].mean()
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
sns.barplot(x='departure_time', y='price', data=db)
plt.title('Price vs Departure Time')

plt.subplot(1,2,2)
sns.barplot(x='arrival_time', y='price', data=db)
plt.title('Price vs Arrival Time')

plt.tight_layout()
plt.show()
```

```python
sns.relplot( x='arrival_time' , y='price' , data=db , col='departure_time' , kind='line')
plt.show()
```
