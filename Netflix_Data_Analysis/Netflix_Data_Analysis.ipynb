Categorizing vote_average column
- we would cut the vote_average value and make 4 categories : popular , average , below_average , not_popular to describe it more using categorize_col() function provide below

```python
#user define function
def categorize_col (df, col, labels):

   edges = [df[col].describe()['min'],
            df[col].describe()['25%'],
            df[col].describe()['50%'],
            df[col].describe()['75%'],
            df[col].describe()['max']]

   df[col] = pd.cut(df[col], edges, labels = labels ,duplicates = 'drop')
   return df
```

```python
labels = ['not_popular','below_aerage','average','popular']
categorize_col(df,'Vote_Average',labels)
df['Vote_Average'].unique()
```
